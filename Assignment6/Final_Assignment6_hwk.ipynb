The task involves Data Cleansing for a text file on Census Cost data using the pandas library.

```
import pandas as pd
```

```
#Use Pandas read_html() function to retrive the HTML tables as dataframes from the web page: https://www.genealogybranches.com/censuscosts.html
df = pd.read_html("https://www.genealogybranches.com/censuscosts.html")
len(df)
#This shows us that there are two data tables in the webpage
```

```
#Exploring the head of the dataset 
df1 = df[0]
df1.head()
```

```
df2 = df[1]
df2.head()
```

```
#We stick to using the first database as it is relevant to our analysis.
df1
```

```
#Defining titles for the columns
df1.iloc[0]
```

```
df1.columns = df1.iloc[0]
```

```
df1.drop([0],inplace=True)
df1.head()
```

```
#Write a function named cleanse_year() that takes a string as input and
#removes any asterisks in the string and return the cleansed string. 
#Test the function by using test strings (for example, "1989*", "20*10")
def cleanse_year(word):
    return word.replace('*', '')

print(cleanse_year('1989*'))
print(cleanse_year('21*10'))
```

```
#Applying the function to the Census Year column.
df1['Census Year'] = df1['Census Year'].apply(cleanse_year)
df1.tail()
```

```
#Creating a function to replace 'Billion' with full numerical values
def cleanse_cost(raw_cost):
    cost = raw_cost.lstrip("$").replace(",","")
    if  raw_cost.find("4.5") != -1:          # if "4.5" in raw_cost:
        cost = "45" + "0" * 8
    elif raw_cost.find("Billion") != -1:     # elif "Billion" in raw_cost: 
        cost = "13" + "0" * 9
    return cost
```

```
#Applying the function to the Census Cost column 
df1['Census Cost'] = df1['Census Cost'].apply(cleanse_cost)
df1.tail()
```

```
print(type(df1['Average Cost Per Person']))
```

```
#Creating a function that replaces the string 'cents' into numerical value with 0s.
def cleanse_cents(cost):
    if "$" in cost:
      cost = cost.replace("$","")
    else:
      cost = cost.strip("cents")
      cost = float(cost)/100
    return cost
```

```
#Testing the pure_dollar function 
print(cleanse_cents('3.10 cents'))
```

```
#Applying the pure_dollar function to the Average Cost Per Person Column. 
df1['Average Cost Per Person']=df1['Average Cost Per Person'].apply(cleanse_cents)
df1
```

```
#Saving the clean dataframe as a csv file.
df1.to_csv('census_cost_clean.csv')
print(df1)
```
